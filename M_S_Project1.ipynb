# Show number of players and their countries

```python
g1 = df.groupby("nationality")['short_name'].size().reset_index(name='number of players')
print(g1)
```

# Show number of players and their top 10 countries


```python
num = df.groupby('nationality')['short_name'].count().head(10)
print(num)
```

# Create a bar plot of top 5 countries and their players

```python
pd.crosstab(df['nationality'].head(5),df['short_name'],normalize='index').plot.bar(stacked=True)
```

# Show top 5 players short name and wages

```python
name_wage = df.groupby('short_name')['wage_eur'].sum().head(5)
print(name_wage)
```

```python
name_wage2 = df.groupby(['short_name'])[['wage_eur']].agg(np.max).head(5)
print(name_wage2)
```

# Show top 5 players short name and wages that are getting highest salaries

```python
df.groupby('short_name')['wage_eur'].max().nlargest(5)
```

# Create a bar plot of previous one

```python
df.groupby('short_name')['wage_eur'].max().nlargest(5).plot.bar(stacked=True)
```
